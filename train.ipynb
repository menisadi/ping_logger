```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(12,5)})
%load_ext autoreload
%autoreload 2
```

```python
import datetime
```

```python
df = pd.read_csv("pings.csv")

# Convert the time column to datetime
df["time"] = pd.to_datetime(df["time"])
```

```python
df['first_train_time'] = (datetime.time(19, 00) > df['time'].dt.time) & (df['time'].dt.time > datetime.time(17, 27))
```

```python
df['second_train_time'] = (datetime.time(20, 00) > df['time'].dt.time) & (df['time'].dt.time > datetime.time(18, 27))
```

```python
df['lost'] = df['delay'] == np.inf
```

```python
df['date'] = df['time'].dt.date
```

```python
office_days = df[(df['time'].dt.dayofweek.isin([0, 2, 3]))]
```

```python
office_days.groupby(['date', 'first_train_time'])[['lost']].mean().plot()
```

```python
office_days.groupby(['date', 'second_train_time'])[['lost']].mean().plot()
```

```python
good_office_days = office_days[office_days['delay'] < np.inf]
```

```python
good_office_days.groupby(['date', 'first_train_time'])[['delay']].mean().plot()
```

```python
good_office_days.groupby(['date', 'second_train_time'])[['delay']].mean().plot()
```

```python
first_train_time_times = office_days.groupby(['date', 'first_train_time'])[['lost']].mean().reset_index()
first_train_time_times = first_train_time_times[first_train_time_times['first_train_time']]
first_train_time_times =first_train_time_times.drop('first_train_time', axis=1)
```

```python
second_train_time_times = office_days.groupby(['date', 'second_train_time'])[['lost']].mean().reset_index()
second_train_time_times = second_train_time_times[second_train_time_times['second_train_time']]
second_train_time_times =second_train_time_times.drop('second_train_time', axis=1)
```

```python
compare_train_times = first_train_time_times.merge(second_train_time_times, on='date', suffixes=('_first', '_second'))
```

```python
compare_train_times.plot()
```

```python
first_train_time_delays = good_office_days.groupby(['date', 'first_train_time'])[['delay']].mean().reset_index()
first_train_time_delays = first_train_time_delays[first_train_time_delays['first_train_time']]
first_train_time_delays =first_train_time_delays.drop('first_train_time', axis=1)
```

```python
second_train_time_delays = good_office_days.groupby(['date', 'second_train_time'])[['delay']].mean().reset_index()
second_train_time_delays = second_train_time_delays[second_train_time_delays['second_train_time']]
second_train_time_delays =second_train_time_delays.drop('second_train_time', axis=1)
```

```python
compare_train_delays = first_train_time_delays.merge(second_train_time_delays, on='date', suffixes=('_first', '_second'))
```

```python
compare_train_delays.plot()
```
